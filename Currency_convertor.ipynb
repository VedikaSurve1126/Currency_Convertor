```python
import requests
from collections import deque
import pandas as pd

def fetch_currencies():
    response = requests.get("https://open.er-api.com/v6/latest/USD")
    data = response.json()
    return data["rates"]

def list_all_currencies():
    rates = fetch_currencies()

    # Create a DataFrame with currency codes
    df = pd.DataFrame(list(rates.keys()), columns=["Currency Code"])

    countries = [
    "United States", "United Arab Emirates", "Afghanistan", "Albania", "Armenia", "Netherlands Antilles",
    "Angola", "Argentina", "Australia", "Aruba", "Azerbaijan", "Bosnia and Herzegovina", "Barbados",
    "Bangladesh", "Bulgaria", "Bahrain", "Burundi", "Bermuda", "Brunei", "Bolivia", "Brazil", "The Bahamas",
    "Bhutan", "Botswana", "Belarus", "Belize", "Canada", "Congo (DRC)", "Switzerland", "Chile", "China",
    "Colombia", "Costa Rica", "Cuba", "Cape Verde", "Czech Republic", "Djibouti", "Denmark", "Dominican Republic",
    "Algeria", "Egypt", "Eritrea", "Ethiopia", "Eurozone", "Fiji", "Falkland Islands", "Faroe Islands",
    "United Kingdom", "Georgia", "Guernsey", "Ghana", "Gibraltar", "Gambia", "Guinea", "Guatemala", "Guyana",
    "Hong Kong", "Honduras", "Croatia", "Haiti", "Hungary", "Indonesia", "Israel", "Isle of Man", "India",
    "Iraq", "Iran", "Iceland", "Jersey", "Jamaica", "Jordan", "Japan", "Kenya", "Kyrgyzstan", "Cambodia",
    "Kiribati", "Comoros", "South Korea", "Kuwait", "Cayman Islands", "Kazakhstan", "Laos", "Lebanon",
    "Sri Lanka", "Liberia", "Lesotho", "Libya", "Morocco", "Moldova", "Madagascar", "North Macedonia",
    "Myanmar (Burma)", "Mongolia", "Macau", "Mauritania", "Mauritius", "Maldives", "Malawi", "Mexico",
    "Malaysia", "Mozambique", "Namibia", "Nigeria", "Nicaragua", "Norway", "Nepal", "New Zealand", "Oman",
    "Panama", "Peru", "Papua New Guinea", "Philippines", "Pakistan", "Poland", "Paraguay", "Qatar", "Romania",
    "Serbia", "Russia", "Rwanda", "Saudi Arabia", "Solomon Islands", "Seychelles", "Sudan", "Sweden",
    "Singapore", "Saint Helena", "Sierra Leone", "Somalia", "Suriname", "South Sudan", "Sao Tome and Principe",
    "Syria", "Eswatini", "Thailand", "Tajikistan", "Turkmenistan", "Tunisia", "Tonga", "Turkey", "Trinidad and Tobago",
    "Tuvalu", "Taiwan", "Tanzania", "Ukraine", "Uganda", "Uruguay", "Uzbekistan", "Venezuela", "Vietnam",
    "Vanuatu", "Samoa", "Central African CFA franc", "Eastern Caribbean dollar", "Special Drawing Rights (SDR)",
    "West African CFA franc", "CFP franc", "Yemen", "South Africa", "Zambia", "Zimbabwe","xyz"]

    # Manually input the countries
    df["Country"] = countries

    print("Available Currencies:")
    print(df)

    print("To search for currency code of a particular country, enter the country name for the same.")
    print("\n")
    user_input = input("Enter the country name : ")

    print("\n")

    if user_input in df['Country'].values:
      row = df.loc[df['Country'] == user_input]
      print("Here is the Currency code you were looking for:")
      print("\n")
      print(row)
      print("\n")
    else:
      print(f"{user_input} does not exist in the DataFrame.")

def currency_converter(amount, from_currency, to_currency, history_stack):
    if from_currency == to_currency:
        return amount

    rates = fetch_currencies()

    try:
        from_rate = rates[from_currency]
        to_rate = rates[to_currency]

        result = (amount / from_rate) * to_rate
        history_stack.append((amount, from_currency, to_currency, result))
        return result
    except KeyError:
        print("Invalid currency.")
        return None

    except FileNotFoundError:
        return {}

favorite_currencies = deque()
conversion_history = deque()

while True:
    print("\n")
    print("1. Convert Currency")
    print("2. Save Favorite Currency")
    print("3. Load Favorite Currencies")
    print("4. View Conversion History")
    print("5. List All Currencies")
    print("6. Exit")
    print("\n")

    choice = input("Enter your choice: ")
    print("\n")

    if choice == "1":
      amount = float(input("Enter the amount: "))
      from_currency = input("Enter the source currency: ").upper()
      to_currency = input("Enter the target currency: ").upper()

      result = currency_converter(amount, from_currency, to_currency, conversion_history)

      if result is not None:
          print(f"{amount} {from_currency} is equal to {result} {to_currency}")

    elif choice == "2":
        favorite_currency = input("Enter the favorite currency: ").upper()
        favorite_currencies.append(favorite_currency)  # Add to the deque
        print(f"Favorite currency {favorite_currency} saved successfully.")

    elif choice == "3":
        print("Favorite Currencies:")
        for currency in favorite_currencies:
            print(currency)

    elif choice == "4":
        print("Conversion History:")
        for entry in conversion_history:
            amount, from_currency, to_currency, result = entry
            print(f"{amount} {from_currency} => {result} {to_currency}")

    elif choice == "5":
        list_all_currencies()

    elif choice == "6":
        break

    else:
        print("Invalid choice. Please try again.")
```
